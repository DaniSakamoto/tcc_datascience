<h2><center>Verificação com Pycaret</center></h2>

Uma vez extraídos os face embeddings do dataset, utilizamos a biblioteca Pycaret para fazer a comparação entre os diversos modelos disponíveis para classificação.

```python
from pycaret.classification import *
import pandas as pd
```

```python
#load the data
import numpy as np
data = np.load('adience-embeddings-2.npz')
trainX, trainy, testX, testy = data['arr_0'], data['arr_1'], data['arr_2'], data['arr_3']
```

```python
#transforma em DataFrame
data = pd.DataFrame(trainX)
```

```python
data['target'] = pd.DataFrame(trainy)
```

```python
data.head()
```

```python
#criação do experimento com o Pycaret
expmclf101 = setup(data = data, target = 'target', session_id = 123)
```

```python
#comparação de performance entre os diversos modelos disponiveis
compare_models()
```

```python
#criação do modelo xgboost
xgboost = create_model('xgboost')
```

```python
#hiperparametrização do xgboost - resultado foi inferior ao original
tuned_xgboost = tune_model(xgboost)
```

```python
#criação de modelo knn
knn = create_model('knn')
```

```python
#hiperparametrização do modelo knn
tuned_knn = tune_model(knn)
```

```python
#criação de modelo extra tree classifier
xtreeclf = create_model('et')
```

```python
#hiperparametrização do modelo extra tree classifier
tuned_xtreeclf = tune_model(xtreeclf)
```
